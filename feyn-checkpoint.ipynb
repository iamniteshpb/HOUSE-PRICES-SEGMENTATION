### Demographic Clustering

```python
from sklearn.cluster import KMeans
```

```python
data.shape
```

```python
X1=data.loc[:,['city','price']].values

sse=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(X1)
    sse.append(kmeans.inertia_)  #inertia is used to find sum of square root error
    
plt.figure()
plt.grid()
plt.plot(range(1,11),sse)
plt.xlabel("K Value")
plt.ylabel("sse")
plt.show()
```

```python
kmeans=KMeans(n_clusters=4)
label=kmeans.fit_predict(X1)
print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(X1[:,0], X1[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],color='black')

plt.xlabel('city')
plt.ylabel('price')
plt.show()
```

### Profiling Segments

```python
data['clusters'] = kmeans.labels_
data['clusters'].size
```

```python
pd.Series.sort_index(data['clusters'].value_counts())
```

```python
size = pd.concat([pd.Series(data['clusters'].size), pd.Series.sort_index(data['clusters'].value_counts())])
```

```python
seg_size = pd.DataFrame(size, columns=['seg_size'])
seg_pct = pd.DataFrame(size/data['clusters'].size, columns=['seg_pct'])
```

```python
seg_pct.T
```

```python
import numpy as np
average = data.apply(np.mean).T
average
```

```python
cluster_average = data.groupby('clusters').apply(np.mean).T
cluster_average
```

```python
profiling_output = pd.concat([average, cluster_average], axis=1)
profiling_output
```

```python
profiling_output_final = pd.concat([seg_size.T, seg_pct.T, profiling_output], axis=0)
```

```python
profiling_output_final.columns = ['Overall', 'Cluster 1', 'Cluster 2','Cluster 3','Cluster 4','Cluster 5']
profiling_output_final
```

```python
profiling_output_final.to_csv('profiling_output_.csv')
```
